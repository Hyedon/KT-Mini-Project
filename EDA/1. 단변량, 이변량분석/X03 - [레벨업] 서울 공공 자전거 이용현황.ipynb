## 3.데이터 전처리

전처리 과정에서 다음과 같은 처리를 할 수 있습니다.

- 결측치 처리
- 값 변경
- 열 추가
- 불필요한 열 제거
- 열 이름 변경
- 데이터 통합(연결 또는 조인)
- 가변수화 등

**1) 결측치 제거**

- 운동량, 탄소량 열은 결측치가 있는 행이 동일하고 몇 건 안되니 제거합니다.

```python
# 결측치 제거
bike.dropna(subset=['운동량'], axis=0, inplace=True)
bike.dropna(subset=['탄소량'], axis=0, inplace=True)

# 확인
print(bike.isna().sum())
```

**2) 결측치 채우기**

- 성별 열은 결측치가 너무 많습니다. 
- 성별별 분석 시 '남','녀' 외에 '기타'로 분석되게 'X'로 채웁니다.

```python
# 결측치 채우기
bike['성별'].fillna('X', inplace=True)

# 확인
bike.isna().sum()
```

**3) 데이터 변경**

- 성별이 소문자 m. f인 값을 대문자 M. F로 변경합니다.

```python
# 성별 소문자 --> 대문자 변경
bike['성별'] = bike['성별'].map({'m':'M','f':'F','M':'M','F':'F'})
#bike.loc[bike['성별'] == 'm', '성별'] = 'M' bike.loc[bike['성별'] == 'f', '성별'] = 'F'
# 확인
bike['성별'].value_counts()
```

- 연령대 코드를 이해하기 쉽게 10대~80대로 변경합니다.

```python
# 연령대 데이터 변경
bike['연령대코드'] = bike['연령대코드'].map({'AGE_001': '10대','AGE_002': '20대','AGE_003': '30대','AGE_004': '40대','AGE_005': '50대','AGE_006': '60대','AGE_007': '70대','AGE_008': '80대'})
#bike['연령대코드'] = bike['연령대코드'].str[-1] + '0대'
# 확인
bike['연령대코드'].value_counts()
```

**4) 불필요한 열 제거**

- 분석에 큰 의미가 없는 대여일자, 대여소번호 열은 제거합니다.

```python
# 대여일자, 대여소번호 제거
bike.drop(['대여일자', '대여소번호'], axis=1, inplace=True)


# 확인
bike.head()
```

**5) 열 이름 변경**

- 분석에 편의를 위해 열 이름을 간단명료하게 변경합니다.
    - 대여소명 --> 대여소
    - 대여구분코드 --> 대여구분
    - 연령대코드  --> 연령대
    - 이동거리(M) --> 이동거리
    - 이용시간(분) --> 이용시간

```python
# 열 이름 변경
bike.rename(columns={'대여소명':'대여소', '대여구분코드':'대여구분', '연령대코드':'연령대', '이동거리(M)':'이동거리', '이용시간(분)':'이용시간'}, inplace=True)


# 확인
bike.head()
```

## 4.데이터 분석

- 데이터를 집계하고 간단히 시각화하여 분석합니다.
- 연속형 값은 선 그래프, 범주형 값은 막대 그래프로 시각화함이 일반적입니다.

**1) 대여구분별 이용 건수**

- 대여구분별 이용 건수를 막대 그래프로 시각화해 확인합니다.
- 이용건수 합 단위가 크니 1,000으로 나누어 시각화에 사용합니다.

```python
# 대여구분별 이용건수
tmp = bike.groupby(by='대여구분', as_index=False)[['이용건수']].sum()
plt.figure(figsize=(10,5))
plt.title('대여구분별 이용건수')
plt.bar(tmp['대여구분'], tmp['이용건수']/1000)
plt.show()

# 시각화

```

```python
tmp = bike.groupby(by='대여구분', as_index=False)[['이용시간']].sum()
plt.figure(figsize=(10,5))
plt.title('대여구분별 이용시간')
plt.bar(tmp['대여구분'], tmp['이용시간']/1000)
plt.show()
```

```python
tmp = bike.groupby(by='대여구분', as_index=False)[['이동거리']].sum()
plt.figure(figsize=(10,5))
plt.title('대여구분별 이동거리')
plt.bar(tmp['대여구분'], tmp['이동거리']/1000000)
plt.ylabel('(천)km')
plt.show()
```

```python
tmp = bike.groupby(by='대여구분', as_index=False)[['이용비율']].mean()
print(tmp)
# 시각화
plt.figure(figsize=(30,5))
plt.title('대여구분별 이용비율')
plt.bar(tmp['대여구분'], tmp['이용비율'])
plt.show()
```

**2) 대여소별 이용 건수 TOP 10**

- 대여소별 이용 건수 TOP 10을 막대 그래프로 시각화해 확인합니다.

```python
# 대여소별 이용건수 TOP 10
tmp = bike.groupby(by='대여소', as_index=False)[['이용건수']].sum()

tmp = tmp.sort_values(by='이용건수', ascending=False).head(10)
print(tmp)
# 시각화
plt.figure(figsize=(30,5))
plt.title('대여소별 이용건수 TOP10')
plt.bar(tmp['대여소'], tmp['이용건수']/1000)
plt.show()


```

```python
tmp = bike.groupby(by='대여소', as_index=False)[['이용시간']].sum()
tmp = tmp.sort_values(by='이용시간', ascending=False).head(10)
plt.figure(figsize=(30,5))
plt.title('대여소별 이용시간')
plt.bar(tmp['대여소'], tmp['이용시간']/1000)
plt.show()
```

```python
tmp = bike.groupby(by='대여소', as_index=False)[['이동거리']].sum()
tmp = tmp.sort_values(by='이동거리', ascending=False).head(10)
plt.figure(figsize=(30,5))
plt.title('대여소별 이동거리')
plt.bar(tmp['대여소'], tmp['이동거리']/1000000)
plt.ylabel('(천)km')
plt.show()
```

```python
bike['이용비율'] = bike['이용시간']/bike['이용건수']
tmp = bike.groupby(by='대여소', as_index=False)[['이용비율']].mean()
tmp = tmp.sort_values(by='이용비율', ascending=False).head(10)
print(tmp)
# 시각화
plt.figure(figsize=(30,5))
plt.title('대여소별 이용비율 TOP10')
plt.bar(tmp['대여소'], tmp['이용비율'])
plt.show()
```
